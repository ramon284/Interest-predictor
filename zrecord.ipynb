```python
import numpy as np
import sounddevice as sd
import time
from datetime import datetime
from scipy.io import wavfile
from scipy.signal import resample
import random

def play_audio(file, device_index):
    fs, data = wavfile.read(file)
    data = data * 2
    with sd.OutputStream(device=device_index, samplerate=fs, channels=2, dtype='float32', callback=None) as stream:
        sd.play(data, samplerate=fs, device=device_index, blocking=True)
def list_devices():
    devices = sd.query_devices()
    for i, device in enumerate(devices):
        print(f"{i}: {device['name']}")
def log_current_time(text):
    current_time = datetime.now().strftime('%H:%M:%S.%f')
    with open('timelog.txt', 'a') as f:
        f.write(f'{text} {current_time}\n')
log_current_time('starting at: ')
print("Available audio devices:")
list_devices()
vb_cable_index = 11 ## CABLE Input (VB-Audio Virtual C !!!!!!!!!!!!!!!!!!

file = 'xylo2.wav'  # Replace this with the path to your audio file
time.sleep(60)

while True:
    sleeptime = random.randint(57, 90)
    log_current_time('sound played at: ')
    play_audio(file, vb_cable_index)    
    time.sleep(sleeptime)
```
